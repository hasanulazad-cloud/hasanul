```python
# Task 1: Data Setup & Display
job_data = {
    "Job_ID": [1, 2, 3, 4, 5],
    "Role": ["Data Analyst", "Data Scientist", "Business Analyst", "Data Engineer", "ML Engineer"],
    "Company": ["ABC Corp", "XyZ Ltd", "DataWorks", "InfraTech", "AI Labs"],
    "Skills": ["Excel SQL Python", "Python ML Statistics", "Excel SQL PowerBI", "Python SQL Spark", "Python ML DeepLearning"]
}

# 1.Print in a readable format
print("--- Current Job Postings ---")
for i in range(len(job_data["Job_ID"])):
    print(f"ID: {job_data['Job_ID'][i]} | Role: {job_data['Role'][i]} | Company: {job_data['Company'][i]} | Skills: {job_data['Skills'][i]}")
```

```python
# Task 2: Add New Job Postings
num = int(input("\nHow many job postings do you want to add? "))

for _ in range(num):
    # Auto-increment Job_ID
    new_id = max(job_data["Job_ID"]) + 1

    # Collect inputs
    role = input("Enter role: ")
    company = input("Enter company: ")
    skills = input("Enter skills: ")

    # Add to dictionary
    job_data["Job_ID"].append(new_id)
    job_data["Role"].append(role)
    job_data["Company"].append(company)
    job_data["Skills"].append(skills)
```

```python
# 3) Clean & Normalize Skills
import re
def clean_skills(skills_string):

    # Convert text to lowercase
    skills = skills_string.lower()

    # Replace commas with spaces
    skills = skills.replace(',', ' ')

    # Optionally remove punctuation
    # This regex removes characters that aren't letters, numbers, or spaces
    skills = re.sub(r'[^\w\s]', '', skills)

    # Remove extra spaces and normalize whitespace
    # split() without arguments splits by any whitespace, join() puts it back with single spaces
    skills = " ".join(skills.split())

    return skills

# Apply clean_skills() to every entry and overwrite the list
job_data['Skills'] = [clean_skills(skill) for skill in job_data['Skills']]

# Print to verify the transformation
print("Cleaned Skills List:")
for skill in job_data['Skills']:
    print(f"'{skill}'")
```

```python
#Task 4: Skill Extraction & Frequency
all_skills = []
for s in job_data["Skills"]:
    all_skills.extend(s.split()) # Create consolidated list

skill_count = {}
for skill in all_skills:
    # Build frequency dictionary
    skill_count[skill] = skill_count.get(skill, 0) + 1

# Sort and print top 5
sorted_skills = sorted(skill_count.items(), key=lambda x: x[1], reverse=True)
print("\nTop skills by frequency:")
for s, c in sorted_skills[:5]:
    print(f"{s}: {c}")
```

```python
#Task 5: Basic Statistics
print(f"\nTotal Jobs: {len(job_data['Job_ID'])}")
print(f"Unique skills: {len(set(all_skills))}") #

# Find Python jobs
python_jobs = [job_data["Role"][i] for i in range(len(job_data["Role"])) if "python" in job_data["Skills"][i]]
print(f"Jobs mentioning Python: {python_jobs}")
```

```python
#Task 6: Skill Search Function
def find_jobs_by_skill(skill_query):
    query = skill_query.lower().strip()
    matches = []
    for i in range(len(job_data["Skills"])):
        if query in job_data["Skills"][i]:
            matches.append((job_data["Job_ID"][i], job_data["Role"][i], job_data["Company"][i]))
    return matches

# Example search
print(f"\nSearch results for 'sql': {find_jobs_by_skill('sql')}")
```

```python
#Task 7: Lambda & List Comprehension
# Identify skills that appear 2 or more times
common = [s for s, c in skill_count.items() if c >= 2]

# Filter roles
common_roles = [job_data["Role"][i] for i in range(len(job_data["Role"]))
                if any(skill in common for skill in job_data["Skills"][i].split())]
print(f"Roles with common skills: {common_roles}")
```

```python
#Task 8: Unique Skill Set
unique_alphabetical = sorted(list(set(all_skills)))
print(f"\nUnique skills (Alphabetical): {unique_alphabetical}")
```

```python
#Task 9: Save to Files
import csv

try:
    # Save job data to CSV
    with open('job_skills.csv', 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["Job_ID", "Role", "Company", "Skills"])
        for i in range(len(job_data["Job_ID"])):
            writer.writerow([job_data["Job_ID"][i], job_data["Role"][i], job_data["Company"][i], job_data["Skills"][i]])

    # Save top skills to TXT
    with open('top_skills.txt', 'w') as f:
        for s, c in sorted_skills:
            f.write(f"{s}: {c}\n")
    print("\nFiles saved successfully!")

except IOError: # Handle errors
    print("Error saving files.")
finally:
    print("Process complete.")
```
